# Chapter 14. Active Galaxies

### Normal galaxy

- Emissions dominated by *stars*

- Stars
    - Hydrodynamic equilibrium
    - Black-body radition
        - Temperature dependent
            - 3,000 K $\lesssim$ T $\lesssim$ 40,000 K
            - Emission wavelength: ~ 4,000 $\mathring{\mathrm{A}}$ to ~ 20,000 $\mathring{\mathrm{A}}$
        

### A small fraction of galaxy is "Active galaxy"
- Much broader SED
    - Emit full wavelength range
        - From X-ray to radio (non-thermal radiation)
- Optical/UV parts of spectra
    - Strong & broad emission lines
- Examples
    - Seyfert galaxies
    - Radio galaxies
    - Quasars
- Emission region: ~ pc scale (very small compared to the host galaxy)
    - $L_\text{AGN} \gg L_\text{host galaxy}$: almost 1,000 times brighter!


### AGN: Important for galaxy formation
- AGN form an important population of galaxies

- AGN are powered by matter accreting onto a SMBH
    - Implies, formation of SMBH is related to formation of galaxy

- Galaxies may have experienced AGN stage

- Feedback: Powerful energy source
    - IGM
    - Formation of galaxies

## 14.1. The Population of Active Galactic Nuclei


### AGN properties

- Compact nuclear region (bright)
- Non-thermal continuum emission
- Strong emission lines
- Variability in continuum / emission lines on relatively short time scales


### (a) Seyfert Galaxies

- Spiral-like morphology with bright star-like nuclei
- Non-thermal continua
- Strong & broad emission lines of high excitation
- Radiation vary ~2 times in ~yr
    - Size of nuclei < 1 light-year
  
#### Two Seyfert types according to the width of emission lines, and it reflect the velocities of the emitting gas clouds

#### Seyfert 1

- Permitted lines: broad H line with FWHM ~ 1,000-5,000 $\text{km s}^{-1}$
- Forbidden lines [OIII]: narrow with FWHM ~ 100 $\text{km s}^{-1}$
- Hard X-ray emission
- Strong non-thermal continua than Seyfert 2
- Large difference in FWHM between permitted and forbidden lines
    - Produced in different regions
        - Comes from a broad-line region

#### Seyfert 2

- Narrow both permitted and forbidden lines with FWHM ~ 100 $\text{km s}^{-1}$
- Likely to emanate from the same region
    - Comes from a narrow-line region
    
    
#### LINERs (Low-ionization nuclear emssion line regions)

- Similar to Seyfert 2
- Low-luminosity extension of Seyferts



### (b) Radio Galaxies

- Relatively strong radio emission
    - $P_\text{1.4 GHz} > 2 \times 10^{23} \text{W Hz}^{-1}$ at ~ 1.4 GHz
- 3C survery (Cambridge third radio survey)
    - Almost all host galaxies of radio AGN: ellipticals
- Strong optical emission lines

#### Classification

#### (1) By line width

#### BLRGs

- Broad Line Radio Galaxies
- Radio loud Seyfert 1

#### NLRGs

- Narrow Line Radio Galaxies
- Radio loud Seyfert 2


#### (2) By radio morphology


#### Double-lobed structure

- ~ 100 kpc scale (even Mpc)

#### Jet-like structure

- Transporting the energy
- Rarely symmetric


#### (3) By radio strength

#### Radio-weak galaxy

- Irregular, more concentrated structure


#### (4) Fanaroff & Riley (1974)

#### FR I

- Weaker radio power
- Distance between the two more intense spots on either side of the nucleus (hot spots)
    - Smaller than half the source size
    
    
#### FR II

- Larger hot spots
- Powerful radio sources
    - $P_\text{1.4 GHz} > 4 \times 10^{25} \text{W Hz}^{-1}$

#### (5) By the slope of the spectrum (fitting the flux at ~ 1GHz with power-law $F_\nu \propto \nu^{-\alpha}$)

#### Steep-spectrum

- Extended soruce

#### $\alpha = 0.4$

#### Flat-spectrum

- Compact source



### (c) Quasars & QSOs

- Radio characteristics
    - Similar to the radio sources
    - High spatial resolution
        - Flat spectrum
            - Very compact nuclei (~ $10^{-3}$ arcsec)
            - Elongated structure

- Optical characteristics
    - Unresolvable at $\theta$ ~ 1"
    - Luminous nuclei 
        - $M_B \lesssim -21.5 + 5 \log h$
        - Unusually blue
    - Variable
    - Detect only using optial
        - Radio invisible
    - Spectrum
        - Similar to Seyfert 1
            - QSO: more luminous
            - Seyfert 1: less luminous
    - Host galaxy: hard to detect
    

### (d) BL Lac Objects and OVVs

All called **Blazers**.

#### Optically Violent Variables (OVVs)

- Varability: less than a day
- Relatively strong polarization (optical)
- Variable at other wavelengths
- Emission lines: continuum is faint

#### BL Lacertae (BL Lac)

- Strong radio emitter
- Highly variable in optical & X-ray
- Strongly polarized radio & optical
- No emission line: continuum is bright
- Sepctra: featureless power-law
    - Similar to flat-spectrrum radio quasars
        - BL Lac objects might be associated with elliptical galaxies.

## 14.2. The Supermassive Black Hole Paradigm

- AGN: emits a large amount of energy from a very small region

How?

"SMBHs" + 
- broad/narrow line region (BLR/NLR)
- accretion disk
- jets

### 14.2.1 The Central Engine

- AGN
    - powered by a SMBH
    - energy source: gravitational potential of the central BH
    

#### Consider a simple spherical model

- Luminosity: $L$


- Gas density distribution $\rho(r)$


- Flux

$$ \large F(r) = \frac{L}{4\pi r^2} $$


- Radiation pressure


$$ \large P_\text{rad} (r) = \frac{L}{4\pi r^2 c} = \frac{F}{c} $$



If the gas is ionized
- pressure force due to the scattering of photons by electron


$$ \large \vec{F}_\text{rad} = \sigma_T P_\text{rad} (r) n_e (r) \hat{r} $$


where


- $\sigma_T$: Thomson scattering cross-section


- $n_e (r)$: electron density at radius r


Pressure force must be smaller than the gravitational force on the gas in order for the gas not to be dispersed quickly.


$$ \large |\vec{F}_\text{rad} \leq F_\text{grav} = \frac{GM_\text{BH}\rho(r)}{r^2} $$

This defines the maximum luminosity (the Eddington luminosity)

#### Eddington luminosity


$$ \large L_\text{Edd} \equiv \frac{4\pi G c m_p}{\sigma_T} B_\text{BH} \approx 1.28 \times 10^{46} M_8 \text{ erg s}^{-1} $$


where


- $M_8 \equiv M_\text{BH}/10^8 M_\odot$


- largest possible luminosity that can be achieved by spherical accretion


- minimum central mass required to achieved the $L_\text{Edd}$


$$ \large M_\text{Edd} = 8 \times 10^7 L_{46} M_\odot $$


where


- $L_{46} \equiv L/(10^{46} \text{ erg s }^{-1})$


#### accretion luminosity (powered by the gravitational potential)


$$ \large L  = \frac{G M_\text{BH}}{r} \dot{M}_\text{BH} $$


where


- $\dot{M}_\text{BH}$: mass accretion rate (rate at which mass crosses a shell of radius r)



#### efficiency (at which the rest mass of accreted material is converted into radiation)


$$ \large \epsilon_r \equiv \frac{L}{\dot{M}_\text{BH}c^2} = \frac{1}{2} \frac{r_S}{r} $$


where 


- $r_S = \frac{2GM_\text{BH}}{c^2} \approx 10^{-2} M_8 \text{ light-days }$: Schwarzchild radius

- bulk of continuum radiation originated from $r \sim 5 r_S, \ \epsilon_R \sim 0.1$ (very high efficiency)
    - with $\epsilon_r \sim 0.1$, $\dot{M} \sim 2 M_\odot \text{ yr }^{-1}$, then $L = 10^{46} \text{ erg s}^{-1}$

#### Eddington accretion rate


$$ \large \dot{M}_\text{Edd} = \frac{L_\text{Edd}}{\epsilon_r c^2} \approx 2.2 M_8 \left( \frac{\epsilon_r}{0.1} \right)^{-1} M_\odot \text{ yr}^{-1} $$

- highest possible accretion rate in the simple spherical model


cf) super-Eddington accretion: if mass accretion is non-spherical
